# Netflix Content Analysis
## Understanding Netflix's Content Strategy Through Data

**Data Source:** Kaggle - Netflix Movies and TV Shows  
**Three Questions:**
1. What is the ratio of Movies vs TV Shows on Netflix?
2. Which countries produce the most content?
3. How is content distributed across age ratings?

```python
import pandas as pd
import matplotlib.pyplot as plt



df = pd.read_csv('netflix_titles.csv',dtype=str)
print(df.head(),df.info())
```


1.What is the ratio of Movies vs TV Shows on Netflix?


```python
movie_show = df['type'].value_counts()

ax = movie_show.plot(kind='bar', figsize=(8, 5), color=['steelblue', 'coral'],
                     title='Netflix Content: Movies vs TV Shows')
ax.set_xlabel('')
ax.set_ylabel('Count',fontsize=12, rotation=0, labelpad=20, ha='right')
plt.xticks(rotation=0)
# show num on each bar
for i, v in enumerate(movie_show):
    ax.text(i, v + 50, str(v), ha='center', fontweight='bold')


plt.tight_layout()
plt.show()
```

2.Which countries produce the most content?


```python
country_series = df['country'].dropna().str.split(',').explode().str.strip()
country_count = df['country'].value_counts(ascending=False)[:15]

ax = country_count.plot(kind='barh',figsize=(8, 5),title='Top 15 Countries by Netflix Content')
ax.set_ylabel('Country',fontsize=10, rotation=0, labelpad=20, ha='right')
plt.tight_layout()
```

3.How is content distributed across age ratings?

```python


rating = df['rating'].value_counts()[:11]

ax = rating.plot(kind='bar',figsize=(8, 5),title='Top 11 rating by Netflix Content')
ax.set_xlabel('rating',ha='right',rotation=0)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

**Key Finding:** Netflix primarily targets adult audiences. TV-MA (mature) and TV-14 
together account for over 70% of all content, while children's content (TV-G, G) 
represents less than 5% of the library.
